```python
import pandas as pd

# Đọc file gốc
file_path = 'h:/SentimentAnalystSchool/MidtermExam/DatasetUnder80kWord/AmazonsElectronics/Kaggle/neutral_output_aspect.csv'
df = pd.read_csv(file_path)

print(f"Đã đọc {len(df):,} dòng dữ liệu")
print(f"\nCác cột: {df.columns.tolist()}")
print(f"\nSố dòng có AspectTerm trống/null: {df['AspectTerm'].isna().sum() + (df['AspectTerm'] == '').sum()}")
df.head(10)
```

```python
# Điền 'None' vào các ô AspectTerm trống hoặc null
df['AspectTerm'] = df['AspectTerm'].fillna('None')
df['AspectTerm'] = df['AspectTerm'].replace('', 'None')

# Kiểm tra lại
print(f"Số dòng có AspectTerm = 'None': {(df['AspectTerm'] == 'None').sum()}")
print(f"Số dòng có AspectTerm trống/null còn lại: {df['AspectTerm'].isna().sum()}")
print(f"\nMẫu dữ liệu sau khi xử lý:")
df.head(10)
```

```python
# Lưu file đã xử lý
output_path = 'h:/SentimentAnalystSchool/MidtermExam/DatasetUnder80kWord/AmazonsElectronics/Kaggle/neutral_output_aspect_filledNone.csv'
df.to_csv(output_path, index=False)

print(f"✓ Đã lưu file: {output_path}")
print(f"Tổng số dòng: {len(df):,}")
```
