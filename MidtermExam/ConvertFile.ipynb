## Quads


```python
import ast
import csv
import json

input_file = "./DatasetMain/Dataraw/hotels/quads/dev.txt"
output_file = "./DatasetMain/DatasetConverted/hotels/quads/dev_converted.csv"

with open(input_file, "r", encoding="utf-8") as f, \
     open(output_file, "w", newline="", encoding="utf-8") as csvfile:

    writer = csv.writer(csvfile)
    writer.writerow(["reviewText", "AspectTerm", "Sentiment"])  # header

    for line in f:
        line = line.strip()
        if not line:
            continue

        # T√°ch reviewText v√† ph·∫ßn label
        if "####" in line:
            text, label_str = line.split("####", 1)
            try:
                labels_list = ast.literal_eval(label_str.strip())
            except:
                labels_list = []
        else:
            text = line
            labels_list = []

        # Chu·∫©n h√≥a: n·∫øu labels_list l√† nhi·ªÅu label
        if labels_list and isinstance(labels_list[0], list):
            all_labels = labels_list
        else:
            all_labels = [labels_list] if labels_list else []

        for labels in all_labels:
            # AspectTerm: tr∆∞·ªùng th·ª© 2 n·∫øu kh√°c 'NULL'
            aspect_term = labels[1] if len(labels) >= 2 and labels[1] != 'NULL' else ""
            # Sentiment: ch·ªâ gi·ªØ positive/negative/neutral (tr∆∞·ªùng th·ª© 3)
            sentiment = labels[2] if len(labels) >= 3 else ""
            if sentiment not in ["positive", "negative", "neutral"]:
                sentiment = ""
            writer.writerow([text, aspect_term, sentiment])
```
