```python
# Lấy aspect extractor từ checkpoint hợp lệ
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(
    checkpoint="english",  # hoặc "multilingual"
    auto_device=True
)
```

```python
# Load dataset
df = pd.read_csv("/kaggle/input/amazonelectronics/AmazonsReviewDropColumns.csv")
df['reviewText'] = df['reviewText'].fillna("").astype(str)
df['Sentiment'] = df['Sentiment'].fillna("Neutral").astype(str)
```

```python
# ==========================================
# 4. HÀM TRÍCH XUẤT 1 ASPECT TERM THEO SENTIMENT
# ==========================================
def extract_single_aspect(sentence, target_sentiment):
    """
    Trích xuất 1 aspect duy nhất theo sentiment:
    - Lấy aspect cùng sentiment với target
    - Nếu nhiều: lấy confidence cao nhất
    - Nếu không có: return None
    """
    if not isinstance(sentence, str):
        return None

    # Predict
    results = aspect_extractor.predict([sentence], print_result=False)
    res = results[0]

    aspects = res.get('aspect', [])
    sentiments = res.get('sentiment', [])
    confidences = res.get('confidence', [])

    # Lọc theo sentiment target
    filtered = [
        (term, conf)
        for term, sent, conf in zip(aspects, sentiments, confidences)
        if sent.lower() == target_sentiment.lower()
    ]

    if not filtered:
        return None

    # Lấy aspect có confidence cao nhất
    filtered.sort(key=lambda x: x[1], reverse=True)
    return filtered[0][0]


# ==========================================
# 5. ÁP DỤNG CHO TOÀN BỘ DATASET
# ==========================================
df['AspectTerm'] = df.apply(
    lambda row: extract_single_aspect(row['reviewText'], row['Sentiment']),
    axis=1
)


# ==========================================
# 6. LƯU FILE OUTPUT
# ==========================================
output_path = "/kaggle/working/output_aspect.csv"
df.to_csv(output_path, index=False)

print("✔ DONE! File đã được lưu tại:", output_path)
```
